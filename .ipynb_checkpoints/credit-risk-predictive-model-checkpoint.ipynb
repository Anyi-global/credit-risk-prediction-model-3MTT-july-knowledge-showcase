## About this project

**Credit risk assessment** is the process of evaluating the likelihood that a borrower will default on a financial obligation, such as a loan or credit card balance. It is a critical part of financial decision-making, particularly for lenders, banks, and financial institutions.

## Purpose of Credit Risk Assessment
- Determine borrower creditworthiness

- Set appropriate loan terms (interest rates, limits)

- Minimize defaults and financial losses

- Comply with regulatory requirements (e.g., Basel III)

## About the Dataset
The dataset contains over 32,000 records with 12 features. It represents the `Credit Bureau` dataset of loan borrowers, including those that repaid and those that defaulted. The target variable is "Loan Status", representing the loan defaulters to be predicted.

### Data Dictionary
**Columns Related to the Borrower's Personal Information:** <br>
`person_age`: The age of the borrower in years.

`person_income`: The annual income of the borrower.

`person_home_ownership`: The type of home ownership of the borrower. This is a categorical variable with the following values:

- `RENT`: The borrower is currently renting a property.

- `OWN`: The borrower owns their home outright.

- `MORTGAGE`: The borrower has a mortgage on their home.

- `OTHER`: Other categories of home ownership.

`person_emp_length`: The number of years the borrower has been employed.

**Columns Related to the Loan**: <br>
`loan_intent`: The purpose for which the loan was requested (e.g., EDUCATION, MEDICAL, VENTURE, PERSONAL, DEBTCONSOLIDATION, HOMEIMPROVEMENT).

`loan_grade`: The credit grade assigned to the loan, typically based on the borrower's credit history and risk of default. This is an ordinal categorical variable ranging from A to G, where:

- `Grade A`: Indicates the lowest risk.

- `Grade G`: Indicates the highest risk.

- `loan_amnt`: The total amount of the loan requested by the borrower.

- `loan_int_rate`: The interest rate of the loan.

**Columns Related to Credit History and Loan Status**: <br>
`loan_status`: The target variable, which indicates whether the loan resulted in a default. This is a binary variable with the following values:

- `0`: No default (the loan was repaid successfully).

- `1`: Default (the borrower failed to repay the loan).

`loan_percent_income`: This represents the Debt-to-Income (DTI) ratio, which is the loan amount as a percentage of the borrower's annual income.

`cb_person_default_on_file`: A binary categorical variable indicating whether the borrower has a history of defaulting on a credit product with a credit bureau. The values are:

- `Y`: Yes, a default is on file.

- `N`: No, no default is on file.

`cb_person_cred_hist_length`: The length of the borrower's credit history in years, since their first loan.

```python
## Detecting outliers on numerical columns

# Filter numerical columns for box plots
numerical_features = df.select_dtypes(include=['number']).columns

# Plot box plots for each numerical feature
plt.figure(figsize=(10, len(numerical_features) * 2))

for i, col in enumerate(numerical_features, 1):
    plt.subplot(len(numerical_features), 1, i)
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot for {col}")
    plt.tight_layout()

plt.show()

#the job of enumerate(numerical_features, 1)
#is to loop over the list 'numerical_features' while also keeping track of an index i starting from 1.
```

```python
df.describe()
```

```python
numerical_columns = df[['person_age', 'person_income', 'work_experience', 'loan_percent_income']].columns

# Define the IQR Outlier Removal Function
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = df

for col in numerical_columns:
    df = remove_outliers_iqr(df, col)
```

```python
# creating a function to plot the distribution of the variables
def univariate_analysis(df):
    num_cols = len(df.columns)
    fig, axes = plt.subplots(nrows=(num_cols // 3) + 1, ncols=3, figsize=(18, 4 * ((num_cols // 3) + 1)))
    fig.tight_layout(pad=5.0)

    axes = axes.flatten() # Flatten axes array for easy iteration

    for i, col_name in enumerate(numerical_columns):
        sns.histplot(data=df, x=col_name, kde=True, ax=axes[i]) # We used KDE for a smooth distribution curve
        axes[i].set_title(f'Distribution of {col_name}')
        axes[i].set_xlabel(col_name)
        axes[i].set_ylabel('Frequency')

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.show()

univariate_analysis(df)
```

```python
# let's create a function for categorical analysis
cat_df = df.select_dtypes(include=['object'])

def cat_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*5, col_name, "*"*5)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print()
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()

cat_univariate_analysis(cat_df)
```

## Apply Data Transformation on Training Data

```python
# separate the training set into categorical and numerical variables
X_train_cat = X_train.select_dtypes(include=['object'])
X_train_num = X_train.select_dtypes(include=['number'])
```

```python
# rescaling the training numerical variables using standardization
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train_num_transformed = pd.DataFrame(std_scaler.fit_transform(X_train_num),
                                       columns=std_scaler.get_feature_names_out(),
                                       index=X_train_num.index)
X_train_num_transformed.head()
```

```python
# Encoding categorical variables using LabelEncoder
from sklearn.preprocessing import LabelEncoder

# initializing the label_encoder
label_encoder = LabelEncoder()

# function that encodes all the categorical features at once
def cat_encoded(cat_features):
    for col in cat_features:
        X_train_cat[col] = label_encoder.fit_transform(X_train_cat[col])
    return X_train_cat[col]

cat_features = X_train_cat.columns

cat_encoded(cat_features)

X_train_cat_transformed = pd.DataFrame(X_train_cat)

X_train_cat_transformed
```

```python
X_train_cat_transformed = X_train_cat_transformed.rename(columns={'home_type': 'home_type_encoded', 'loan_intent': 'loan_intent_encoded', 
                                                                  'loan_grade': 'loan_grade_encoded', 'previous_default': 'previous_default_encoded'})
X_train_cat_transformed
```

## Apply Data Preparation on Test Data

```python
# separate the training set into categorical and numerical variables
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])
```

```python
X_test_cat.head()
```

```python
X_test_num.head()
```

```python
# rescaling the training numerical variables using standardization
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_test_num_transformed = pd.DataFrame(std_scaler.fit_transform(X_test_num),
                                       columns=std_scaler.get_feature_names_out(),
                                       index=X_test_num.index)
X_test_num_transformed.head()
```

```python
# Encoding categorical variables using LabelEncoder
from sklearn.preprocessing import LabelEncoder

# initializing the label_encoder
label_encoder = LabelEncoder()

# function that encodes all the categorical features at once
def cat_encoded(cat_features):
    for col in cat_features:
        X_test_cat[col] = label_encoder.fit_transform(X_test_cat[col])
    return X_test_cat[col]

cat_features = X_test_cat.columns

cat_encoded(cat_features)

X_test_cat_transformed = pd.DataFrame(X_test_cat)

X_test_cat_transformed.head()
```

```python
X_test_cat_transformed = X_test_cat_transformed.rename(columns={'home_type': 'home_type_encoded', 'loan_intent': 'loan_intent_encoded', 
                                                                  'loan_grade': 'loan_grade_encoded', 'previous_default': 'previous_default_encoded'})
X_test_cat_transformed
```

```python
# Concatenate X_train_num_transformed and X_train_cat_transformed
X_test_transformed = pd.concat([X_test_num_transformed, X_test_cat_transformed], axis=1)

X_test_transformed.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score, f1_score

## Model Performance Evaluation

# Define models to compare
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "XGB": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(n_neighbors=5)
}

# Store results
results = {}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    results[name] = {
        'accuracy': acc,
        'F1 Score': f1_score(y_test, y_pred),
        'classification_report': report
    }

results
```

## Insight
From the model evaluation, the full classification reports gives a complete view of how each model performs on both classes, especially the minority class (1 = defaulters), which is most critical in credit risk prediction.

### Objective Recap
We want to predict loan default accurately. So we care most about:

- Precision for class 1: how many predicted defaulters were real

- Recall for class 1: how many actual defaulters the model caught

- F1-Score for class 1: balance between precision and recall

- Overall Accuracy: still useful but secondary due to class imbalance

### Best Model: Random Forest
**Why it's the best**:

- Highest F1-score for defaulters: 0.80

- Very high precision: 0.946 â†’ very few false positives

- Solid recall: 0.694 â†’ catches a good number of defaulters

- Best accuracy: 92.92%

Even though Gradient Boosting and XGBoost come close, Random Forest wins overall for:

- Better balance between catching defaulters (recall) and being sure of them (precision)

- Highest F1-score, which is key in imbalanced classification tasks

## Retraining our Best Model and Applying `'Class Weighting'` to increase `'Recall'`

```python
# Collect evaluation metrics for the best model (Random Forest)
from sklearn.metrics import classification_report

# Retraining Random Forest and Applying Class Weighting to increase Recall
best_model = RandomForestClassifier(class_weight='balanced', random_state=42)
best_model.fit(X_train_transformed, y_train)
y_pred = best_model.predict(X_test_transformed)

# Classification report
print(classification_report(y_test, y_pred))
```
